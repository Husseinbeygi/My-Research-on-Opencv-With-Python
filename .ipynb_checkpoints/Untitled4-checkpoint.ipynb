```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("DATA/5.jpg").astype(np.float32) / 255
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
def dp_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
dp_img(img)
```

```python
gamma = 1/3
```

```python
img2 = np.power(img,gamma)
```

```python
dp_img(img2)
```

```python
def txtimg():
    img = cv2.imread("DATA/5.jpg").astype(np.float32) / 255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(img,text='Fly',org=(600,600),fontFace=font,fontScale=10,color=(255,0,0),thickness=10)
    return img
```

```python
dp_img(txtimg())
```

```python
kernel = np.ones((5,5)) / 25
```

```python
dp_img(cv2.filter2D(txtimg(),-1,kernel))
```

```python
dp_img(cv2.blur(txtimg(),ksize=(10,10)))
```
