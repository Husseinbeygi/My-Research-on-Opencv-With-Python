```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("DATA/2.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2 = cv2.imread("DATA/3.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
Rimg2 = cv2.resize(img2,(736,736))
```

```python
plt.imshow(Rimg2)
```

```python
addimg = cv2.addWeighted(img1,0.1,Rimg2,0.2,0.2)
```

```python
plt.imshow(addimg)
```

```python
_,ther1 = cv2.threshold(img1,110,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(ther1,cmap='gray')
```

```python
th2 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
plt.imshow(th2)
```
